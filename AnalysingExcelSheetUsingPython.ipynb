```python
from openpyxl.styles.borders import BORDER_THICK
from openpyxl.styles import Border, Side
from openpyxl.styles import Font
from openpyxl import Workbook
from datetime import datetime
from pycel import ExcelCompiler
from pandas import *

book = Workbook()

sheet = book.active

ws = wb_obj['DATA']
b_col, c_col = ['B', 'C']
d_col, e_col = ['D', 'E']
f_col, g_col = ['F', 'AG']
for row in range(2, 46238):
    #result_cell = 'B{}'.format(row)
    
    b_value = str(ws[b_col + str(row)].value)
    c_value = str(ws[c_col + str(row)].value)
    
    def days_between(d1,d2):
        d1 = datetime.strptime(d1, "%Y-%m-%d %H:%M:%S")
        
        d2 = datetime.strptime(d2, "%Y-%m-%d %H:%M:%S")
        
        return abs((d2 - d1).days)
    
    d_value = ws[d_col + str(row)].value
    e_value = ws[e_col + str(row)].value
    f_value = ws[f_col + str(row)].value
    
    
    def OutStandingBal(OB,CS):
      
        if OB == None:
            OB = 0
        else:
            OB=OB
        x = days_between(b_value,c_value)
        total = 0
        OutBal = 0
        OB = OB
        if (x >= 0 and x <=31):
            OutBal=OB*0
            total =OutBal
            result_cell = A.format(row)
            sheet[result_cell] = total
            sheet[a] = AS
            sheet[at] = T
            sheet[C1] = aT
            
        if x >= 32 and x <= 62:
            OutBal=OB*0.25
            total =OutBal
            result_cell = B.format(row)
            sheet[result_cell] = total
            sheet[b] = BS
            sheet[bt] = T
            sheet[C2] = bT
                    
        elif x >= 63 and x <= 93:
            OutBal=OB*0.5
            total =OutBal
            result_cell = C.format(row)
            sheet[result_cell] = total
            sheet[c] = CS
            sheet[ct] = T
            sheet[C3] = cT
                   
        elif x >= 94 and x <= 124:
            OutBal=OB*0.65
            total =OutBal
            result_cell = D.format(row)
            sheet[result_cell] = total
            sheet[d] = DS
            sheet[dt] = T
            sheet[C4] = dT
            
        elif x >= 125 and x <= 155:
            OutBal=OB*0.85
            total =OutBal
            result_cell = E.format(row)
            sheet[result_cell] = total
            sheet[e] = ES
            sheet[et] = T
            sheet[C5] = eT
                    
        elif (x >= 156) :
            OutBal=OB*1
            total =OutBal
            result_cell = F.format(row)
            sheet[result_cell] = total
            sheet[f] = FS
            sheet[ft] = T
            sheet[C6] = fT
            
                    
    
    #OPA_cell = 'D{}'.format(row)
    #CPA_cell = 'E{}'.format(row)
    OpenDate = "2019-6-30 00:00:00"
    OpenDate = datetime.strptime(OpenDate, "%Y-%m-%d %H:%M:%S")
    CloseDate = "2019-7-22 00:00:00"
    CloseDate = datetime.strptime(CloseDate, "%Y-%m-%d %H:%M:%S")
    date = b_value
    date = datetime.strptime(date, "%Y-%m-%d %H:%M:%S")
    def columnDateOne(da):
        if da <= OpenDate and da < CloseDate:
            a, b, c, d, e, f = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6']
            A, B, C, D, E, F = ['G{}', 'H{}', 'J{}', 'K{}', 'L{}', 'M{}']
            AS, BS, CS, DS, ES, FS = ['=SUM(G1:G46237)', '=SUM(H1:H46237)', '=SUM(J1:J46237)', 
                                      '=SUM(K1:K46237)', '=SUM(L1:L46237)', '=SUM(M1:M46237)']
            OutStandingBal(d_value, e_value)    
        
        elif da <= CloseDate:
            a, b, c, d, e, f = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6']
            A, B, C, D, E, F = ['N{}', 'O{}', 'P{}', 'Q{}', 'R{}', 'S{}']
            AS, BS, CS, DS, ES, FS = ['=SUM(N1:N46237)', '=SUM(O1:O46237)', '=SUM(P1:P46237)', 
                                      '=SUM(Q1:Q46237)', '=SUM(R1:R46237)', '=SUM(S1:S46237)']
            OutStandingBal(d_value, e_value)
            
    def columnDateTwo(da):
        if da <= OpenDate and da < CloseDate:
            a, b, c, d, e, f = ['A8', 'A9', 'A10', 'A11', 'A12', 'A13']
            A, B, C, D, E, F = ['U{}', 'V{}', 'W{}', 'X{}', 'Y{}', 'Z{}']
            AS, BS, CS, DS, ES, FS = ['=SUM(U1:U46237)', '=SUM(V1:V46237)', '=SUM(W1:W46237)', 
                                      '=SUM(X1:X46237)', '=SUM(Y1:Y46237)', '=SUM(Z1:Z46237)']
            OutStandingBal(d_value, e_value)    
        
        elif da <= CloseDate:
            a, b, c, d, e, f = ['B8', 'B9', 'B10', 'B11', 'B12', 'B13']
            A, B, C, D, E, F = ['AA{}', 'AB{}', 'AC{}', 'AD{}', 'AE{}', 'AF{}']
            AS, BS, CS, DS, ES, FS = ['=SUM(AA1:AA46237)', '=SUM(AB1:AB46237)', '=SUM(AC1:AC46237)', 
                                      '=SUM(AD1:AD46237)', '=SUM(AE1:AE46237)', '=SUM(AF1:AF46237)']
            OutStandingBal(d_value, e_value)
    
    
    
    
    ProductType = f_value
    PT_cell = 'AM{}'.format(row)
    PS_cell = 'AN{}'.format(row)
    
    if "TV" in ProductType:
        T = "SHS"
        
        at, bt, ct, dt, et, ft = ['D1', 'D2', 'D3', 'D4', 'D5', 'D6']
        C1, C2, C3, C4, C5, C6 = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6']
        aT, bT, cT, dT, eT, fT = ['0-31', '32-62', '63-93', '94-124', '125-155', '156+']
        
        sheet[PT_cell] = columnDateOne(date)
        
    elif "SHS" in ProductType:
        T = "TV"
        
        at, bt, ct, dt, et, ft = ['D8', 'D9', 'D10', 'D11', 'D12', 'D13']
        C1, C2, C3, C4, C5, C6 = ['C8', 'C9', 'C10', 'C11', 'C12', 'C13']
        aT, bT, cT, dT, eT, fT = ['0-31', '32-62', '63-93', '94-124', '125-155', '156+']
        sheet[PS_cell] = columnDateTwo(date)
 
   
        
book.save('solution.xlsx')
path = 'solution.xlsx'
wb = openpyxl.load_workbook(path)
sht = wb.active 
finalbook = Workbook()
fw = finalbook.active

fw.column_dimensions['A'].width = 20
fw.column_dimensions['B'].width = 30
fw.column_dimensions['C'].width = 40
fw.column_dimensions['D'].width = 40

excel = ExcelCompiler(filename='solution.xlsx')


thin_border = Border(left=Side(border_style='thin', color='000000'),
                     right=Side(border_style='thin', color='000000'),
                     top=Side(border_style='thin', color='000000'),
                     bottom=Side(border_style='thin', color='000000'))

ws.cell(row=1, column=4).border = thin_border

fw["A1"] = "PRODUCT TYPE"
fw["A1"].font = Font(bold=True)
fw["B1"] = "DAYS WITHOUT LIGHT"
fw["B1"].font = Font(bold=True)
fw["C1"] = "OPENING PROVISION AMOUNT"
fw["C1"].font = Font(bold=True)
fw["D1"] = "CLOSING PROVISIONING AMOUNT"
fw["D1"].font = Font(bold=True)
fw["A18"] = "PROVISIONING TOTAL-AMOUNT"
fw["A18"].font = Font(bold=True)

d1 = str(sht['D1'].value)
d2 = str(sht['D2'].value)
d3 = str(sht['D3'].value)
d4 = str(sht['D4'].value)
d5 = str(sht['D5'].value)
d6 = str(sht['D6'].value)
d7 = str(sht['D7'].value)
d8 = str(sht['D8'].value)
d9 = str(sht['D9'].value)
d10 = str(sht['D10'].value)
d11 = str(sht['D11'].value)
d12 = str(sht['D12'].value)
d13 = str(sht['D13'].value)

fw["A2"] = d1
fw["A3"] = d2
fw["A4"] = d3
fw["A5"] = d4
fw["A6"] = d5
fw["A7"] = d6
fw["A8"] = ""
fw["A9"] = d8
fw["A10"] = d9
fw["A11"] = d10
fw["A12"] = d11
fw["A13"] = d12
fw["A14"] = d13

c1 = str(sht['C1'].value)
c2 = str(sht['C2'].value)
c3 = str(sht['C3'].value)
c4 = str(sht['C4'].value)
c5 = str(sht['C5'].value)
c6 = str(sht['C6'].value)
c7 = str(sht['C7'].value)
c8 = str(sht['C8'].value)
c9 = str(sht['C9'].value)
c10 = str(sht['C10'].value)
c11 = str(sht['C11'].value)
c12 = str(sht['C12'].value)
c13 = str(sht['C13'].value)

fw["B2"] = c1
fw["B3"] = c2
fw["B4"] = c3
fw["B5"] = c4
fw["B6"] = c5
fw["B7"] = c6
fw["B8"] = ""
fw["B9"] = c8
fw["B10"] = c9
fw["B11"] = c10
fw["B12"] = c11
fw["B13"] = c12
fw["B14"] = c13

fw["C2"] = excel.evaluate('Sheet!A1')
fw["C3"] = excel.evaluate('Sheet!A2')
fw["C4"] = excel.evaluate('Sheet!A3')
fw["C5"] = excel.evaluate('Sheet!A4')
fw["C6"] = excel.evaluate('Sheet!A5')
fw["C7"] = excel.evaluate('Sheet!A6')

fw["C9"] = excel.evaluate('Sheet!A8')
fw["C10"] = excel.evaluate('Sheet!A9')
fw["C11"] = excel.evaluate('Sheet!A10')
fw["C12"] = excel.evaluate('Sheet!A11')
fw["C13"] = excel.evaluate('Sheet!A12')
fw["C14"] = excel.evaluate('Sheet!A13')

fw["D2"] = excel.evaluate('Sheet!B1')
fw["D3"] = excel.evaluate('Sheet!B2')
fw["D4"] = excel.evaluate('Sheet!B3')
fw["D5"] = excel.evaluate('Sheet!B4')
fw["D6"] = excel.evaluate('Sheet!B5')
fw["D7"] = excel.evaluate('Sheet!B6')

fw["D9"] = excel.evaluate('Sheet!B8')
fw["D10"] = excel.evaluate('Sheet!B9')
fw["D11"] = excel.evaluate('Sheet!B10')
fw["D12"] = excel.evaluate('Sheet!B11')
fw["D13"] = excel.evaluate('Sheet!B12')
fw["D14"] = excel.evaluate('Sheet!B13')


fw["C18"] = "=SUM(C2:C14)"
fw["D18"] = "=SUM(D2:D14)"
fw["E18"] = "=D8-C18"



finalbook.save('FinalReport.xlsx')

path = 'FinalReport.xlsx'


xls = ExcelFile(path)
df = xls.parse(xls.sheet_names[0])
print (df.to_dict())
```
